# Run `experiments/compare.yaml` (Colab or local Jupyter)

This notebook runs the PAN vs Transformer comparison experiment and previews the key outputs.


## 1) Environment setup

- **Colab**: update `REPO_URL` and `REPO_DIR`, then run this cell to clone/install.
- **Local Jupyter**: keep `REPO_DIR='.'` if your notebook server is opened at repo root.


```python
from pathlib import Path
import os
import subprocess
import sys

IN_COLAB = 'google.colab' in sys.modules
REPO_URL = 'https://github.com/<your-org>/<your-repo>.git'  # fill in for Colab
REPO_DIR = Path('/content/phase-accumulator-network-research' if IN_COLAB else '.').resolve()

if IN_COLAB and not REPO_DIR.exists():
    subprocess.run(['git', 'clone', REPO_URL, str(REPO_DIR)], check=True)

# Install editable package + notebook deps
subprocess.run([sys.executable, '-m', 'pip', 'install', '-e', str(REPO_DIR), 'ipykernel'], check=True)
print('Using repository:', REPO_DIR)
```

## 2) Run the experiment

This executes `python -m pan_lab experiments/compare.yaml`, writing outputs to `results/compare/`.


```python
from pathlib import Path
import subprocess
import sys

repo_dir = REPO_DIR
yaml_path = repo_dir / 'experiments' / 'compare.yaml'
assert yaml_path.exists(), f'Missing experiment file: {yaml_path}'

subprocess.run([sys.executable, '-m', 'pan_lab', str(yaml_path)], cwd=repo_dir, check=True)
```

## 3) Preview outputs


```python
from pathlib import Path
import pandas as pd

out_dir = REPO_DIR / 'results' / 'compare'
runs_csv = out_dir / 'runs.csv'
curves_csv = out_dir / 'curves.csv'

print('Output directory:', out_dir)
print('Exists:', out_dir.exists())
print('Files:', sorted([p.name for p in out_dir.glob('*')])[:20])

if runs_csv.exists():
    runs_df = pd.read_csv(runs_csv)
    display(runs_df[['run_id', 'model_kind', 'label', 'final_val_acc', 'peak_val_acc', 'param_count']])
else:
    print('runs.csv not found yet.')

if curves_csv.exists():
    curves_df = pd.read_csv(curves_csv)
    display(curves_df.head())
else:
    print('curves.csv not found yet.')
```
